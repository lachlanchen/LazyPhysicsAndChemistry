```python
import sympy as sym
```

# X

```python
H, h, u = sym.symbols('H, h, u')

x_ab, x_cd, x_ef, p_ab, p_cd, p_ef, z_ac, z_ae = sym.symbols(
    'X_{AB}, X_{CD}, X_{EF}, P_{AB}, P_{CD}, P_{EF}, Z_{AC}, Z_{AE}'
)

eq1 = sym.Eq(H/x_ab, h/(p_ab*u))
eq2 = sym.Eq((H+z_ac)/x_cd, h/(p_cd*u))
eq3 = sym.Eq((H+z_ae)/x_ef, h/(p_ef*u))

# eq1 = sym.Eq(H*(p_ab*u)-x_ab*h, 0)
# eq2 = sym.Eq((H+z_ac)*(p_cd*u)-x_cd*h, 0)
# eq3 = sym.Eq((H+z_ae)*(p_ef*u)-x_ef*h, 0)

result1 = sym.solve([eq1, eq2],(H, h))
```

```python
eq1, eq2, result1
```
