```python
from numerov import find_eigen
import matplotlib.pyplot as plt
```

```python
# def f(current, E):
#      return -2*(1/current+E)
    
def lennard_jones(current, E, l=5):
        
#     if current < 10:
    
#         V = 5.9*(( 1/current)**12 - 2*(1/current)**6)
#         # V = 0 if V >=0 else V
#     else:
#         V=0

    V = 5.9 * (( 3.57/current)**12 - 2*(3.57/current)**6)
    L = l * (l+1) / (6.12 * current**2)
    
    return -6.12*(-V + L + E)
    
    
```

```python
eigen_guessed, wave_function = find_eigen(lennard_jones, 
                                          E_lower=0, E_upper=10, n_levels=100, 
                                          current=10*3.57, start=0.8*3.57, end=100, 
                                          x_current=0, dx_current=0.07, 
                                          direction=-1, n_steps=10000)
```

```python
rv, xhv = wave_function[0]
plt.plot(rv[:], xhv[:])
```
